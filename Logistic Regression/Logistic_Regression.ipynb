```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=load_iris()
data
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
df['target'] = data.target
```

```python
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(2)
```

```python
df.target.unique()
```

```python
#For Binary Classification
df_final = df[df['target'] != 2]
```

```python
df_final.target.unique()
```

```python
#Divide into X and y
X = df_final.iloc[:, :-1]
y = df_final.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
#Train Test Data Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
X_train.shape, X_test.shape
```

```python
X_train
```

```python
#Model Training
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
#Corresponding to higher probability class is predicted, cutoff 0.5
classifier.predict_proba(X_test)
#Output gives probability in form of pairs
# [Probability of class 0, Probability of class 1]
```

```python
#Evaluation metrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
# 1 means 100%
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Macro avg: For each class, take simple avg
# Avg precison, recall, f1 score, support of class 0 and class 1 will be macro avg
# Weighted avg is used in case of class imbalance, it gives importance to performance on dominant class
# Support means, number of data points of that class
```

```python
from sklearn.metrics import roc_curve, auc

#Get predicted probablilty for class 1:
classifier.predict_proba(X_test)[:, 1]
```

```python
y_pred_proba = classifier.predict_proba(X_test)[:, 1]
```

```python
# Reciever operating characteristics curve
# Inputs: y_test, y_pred_proba. 
# It will return TPR, FPR with different cutoff of probabilities
# It shows how well your model is able to distinguish between two classes
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
```

```python
fpr
```

```python
tpr
```

```python
thresholds
```

```python
#Calculate AUC score
roc_auc = auc(fpr, tpr)
roc_auc
```

```python
#Plot ROC-AUC curve

plt.figure(figsize = (8, 6))
#Plot ROC curve
plt.plot(fpr, tpr, color = 'darkorange', linewidth = 2, label = 'ROC curve (area = %0.2f' % roc_auc)
#Plot diagonal line representing random guessing
plt.plot([0,1], [0,1], color='navy', linewidth = 2, linestyle = '--')

plt.xlim([0.0, 1.0]) #X-axis limit
plt.ylim([0.0, 1.05]) #Y-axis limit

plt.xlabel('False Positive Rate')
plt.ylabel('True Psoitive Rate')

plt.title('Reciever Operating Characteristics (ROC) curve')
plt.legend(loc = 'lower right')
plt.show()
```
