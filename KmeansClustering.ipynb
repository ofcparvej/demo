```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples = 1000, centers = 3, n_features = 2, random_state = 1)
```

```python
X
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans
```

```python
kmeans.fit(X_train)
```

```python
y_labels = kmeans.predict(X_train)
```

```python
y_labels
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_labels) #On train data, y_labels is predicted on train data
```

```python
#Checking performance of clustering model:
from sklearn.metrics import silhouette_score
silhouette_score(X_train, kmeans.labels_)
#Closer to 1 -> Good model
```

```python
kmeans.labels_
```

```python
kmeans.inertia_ #k=3, this is wcss distance
```

```python
#Elbow method to select the value of K:
wcss = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
#Plot elbow curve:
plt.plot(range(1, 10), wcss)
plt.xticks(range(1, 10))
plt.xlabel("No. of cluster")
plt.ylabel("WCSS")
plt.show()
```

```python
#Based on the plot, no. of clusters should be 3
```
