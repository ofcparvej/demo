# Simple Linear Regression
#### Simple linear regression is a regression model that estimates the relationship between one independent variable and one dependent variable using a straight line.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
diabetes
```

```python
print(diabetes.DESCR)
```

## Problem Statement:
#### To predict disease progression 1 year after baseline given 10 10 features (age, sex, bmi, bp, s1, s2, s3, s4, s5, s6)

```python
diabetes.data
```

```python
diabetes.target
```

```python
diabetes.feature_names
```

```python
df=pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
df['target'] = diabetes.target
```

```python
df
```

```python
#Next step is EDA, Data Cleaning, Data Preparation and Feature Engineering
#But since the data is already clean, we can skip this step for this example
```

```python
#Divide the data into dependent(Y) and independent(X) variables
#Since this is simple linear regression, we will take only one random feature

X=df[['bmi']]
y=df['target']
```

```python
X
```

```python
y
```

```python
#Split dataset into training and testing data
from sklearn.model_selection import train_test_split
```

```python
# test_size means % data for testing, eg: 0.2 means 20% data will be testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Scaling 
# The data is already scaled so this step can be skipped for this example
```

```python
#Model training
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
#Train the model using training data
model.fit(X_train, y_train)
#This will use gradient descent to calculate optimal coefficients
```

```python
#Let's see the coefficients:
model.coef_
```

```python
model.intercept_
```

```python
#Predicting on test data
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#Visualize the result:
plt.scatter(X_test, y_test, color='black', label='Actual Data')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Linear Regression Line')
plt.xlabel('BMI')
plt.ylabel('One year progression target')
plt.legend()
plt.show()
```

```python
#Evaluation Metrices

from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```
