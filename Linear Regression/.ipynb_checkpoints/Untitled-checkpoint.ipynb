```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generating synthetic data points

np.random.seed(1)
```

```python
#Generating synthetic X

X = 2 * np.random.rand(100, 1)
```

```python
X
```

```python
#Generating synthetic data points y

y = 4 + 3*X + 1.5*X**2 + np.random.randn(100,1)
```

```python
y
```

```python
#Split into testing and trining data

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
#Perform ploynomial regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
degree = 2
poly_features = PolynomialFeatures(degree = 2, include_bias = False)
X_poly_train = poly_features.fit_transform(X_train)
```

```python
X_poly_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg
```

```python
reg.fit(X_poly_train, y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#Making prediction on training data itself

y_poly_predict = reg.predict(X_poly_train)
```

```python
y_poly_predict
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse_train = mean_squared_error(y_train, y_poly_predict)
print(f'Mean Squared Error on training data: {mse_train}')
```

```python
plt.scatter(X_train, y_train, label = "Training Data")
```

```python
plt.scatter(X_train, y_train, label="Training Data")
X_range = np.linspace(0, 2, 100).reshape(-1, 1)
X_range_poly = poly_features.transform(X_range)
plt.plot(X_range, reg.predict(X_range_poly), color='red', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
